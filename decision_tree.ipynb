```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
df
```

```python
y=df['salary_more_then_100k']
x=df.drop(['salary_more_then_100k'],axis='columns')
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
x['company']=le_company.fit_transform(x['company'])
```

```python
x['job']=le_job.fit_transform(x['job'])
```

```python
x['degree']=le_degree.fit_transform(x['degree'])
```

```python
x
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model_dt.fit(xtrain,ytrain)
```

```python
ypred=model_dt.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy score =",accuracy_score(ypred,ytest))
```
