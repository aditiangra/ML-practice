```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
```

```python
df.info()
```

```python
df
```

```python
y=df['Class Type']
x=df.drop('Class Type',axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
new_x=scalar.fit_transform(x)
new_x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(new_x,y,test_size=0.2,random_state=42)
```

```python
df.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier(max_depth=5,n_estimators=10)
```

```python
model_rf.fit(xtrain,ytrain)
```

```python
ypred=model_rf.predict(xtest)
ypred
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy score is ",accuracy_score(ytest,ypred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
color='white'
matrix=plot_confusion_matrix(model_rf, xtest, ytest, cmap=plt.cm.Blues)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
