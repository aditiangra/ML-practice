#### Bagging (Bootstrap Aggregating) 

Given a standard original data set D of size n, bagging generates m new data sets D1, D2...Dm, each of size less than n, with replacement from an original sample. 
A bootstrap sample may have a few duplicate observations or records, as the sampling is done with replacement.

Train the models on new training sets of the original dataset and then aggregate their individual predictions (either by voting or by averaging) to form a final prediction
