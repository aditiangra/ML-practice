```python
#importing all the libraries io,pd,np,plt,linear_reg
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#making the dataframe
df = pd.read_excel('data.xlsx')
```

```python
df
```

```python
%matplotlib inline
#This is used to plot the graph and merge backend and frontend
plt.xlabel('Price')
plt.ylabel('Area')
plt.scatter(df.area,df.price,color='red')
#plot(x,y,color)
```

```python
#features and traget needed to be converted in numpy array of shape(-1,1) hence rehsaped because .fit takes this
y=np.array(df['price']).reshape(-1,1)
x=np.array(df['area']).reshape(-1,1)
```

```python
y
```

```python
x
```

```python
#model trained
model_lm=LinearRegression()
model_lm.fit(x,y)
```

```python
model_lm.predict([[3300]])
```

```python
model_lm.coef_
```

```python
model_lm.intercept_
```
