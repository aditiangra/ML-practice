```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
```

```python
df.info()
```

```python
df
```

```python
y=df['Class Type']
x=df.drop('Class Type',axis=1)
```

```python

y
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
new_x=scalar.fit_transform(x)
new_x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(new_x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
knn_model=KNeighborsClassifier(11,algorithm='brute')
```

```python
model_nb=GaussianNB()
```

```python
knn_model.fit(xtrain,ytrain)
```

```python
model_nb.fit(xtrain,ytrain)
```

```python
y_predict_model_nb=model_nb.predict(xtest)
```

```python
y_predict_knn=knn_model.predict(xtest)
```

```python
y_predict_knn
```

```python
y_predict_model_nb
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy score =",accuracy_score(y_predict_knn,ytest))
```

```python
from sklearn.metrics import accuracy_score
print("The accuracy score =",accuracy_score(y_predict_model_nb,ytest))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest, y_predict_knn)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest, y_predict_model_nb)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
color='white'
matrix=plot_confusion_matrix(knn_model, xtest, ytest, cmap=plt.cm.Blues)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
color='white'
matrix=plot_confusion_matrix(model_nb, xtest, ytest, cmap=plt.cm.Blues)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.metrics import classification_report
 
print(classification_report(ytest, y_predict_knn))
```

```python
from sklearn.metrics import classification_report
 
print(classification_report(ytest, y_predict_model_nb))
```
