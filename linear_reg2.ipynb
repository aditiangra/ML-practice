```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
```

```python
df
```

```python
df.rename(columns = {'per capita income (US$)':'income'}, inplace = True)
```

```python
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df.info()
```

```python
x=np.array(df['year']).reshape(-1,1)
y=np.array(df['income']).reshape(-1,1)
```

```python
#model trained
model_lm=LinearRegression()
```

```python
model_lm.fit(x,y)
```

```python
model_lm.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income,color='red')
plt.plot(df.year,model_lm.predict(x),color='blue')
```
